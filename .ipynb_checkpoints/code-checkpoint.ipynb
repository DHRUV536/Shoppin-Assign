## 2. CNN Based Feature Extraction 
* Using Pretrained Model to extract features from images.
* Using Cosine_similarity to find similar feature maps.

```python

# Load Pre-trained CNN Model (ResNet50)
def load_model():
    base_model = ResNet50(weights='imagenet', include_top=False, pooling='avg')
    model = Model(inputs=base_model.input, outputs=base_model.output)
    return model

# Preprocess Image
def preprocess_image(image_path, target_size=(224, 224)):
    image = load_img(image_path, target_size=target_size)
    image_array = img_to_array(image)
    image_array = np.expand_dims(image_array, axis=0)
    return preprocess_input(image_array)

# Extract Features
def extract_features(image_paths, model):
    features = {}
    for image_path in image_paths:
        image = preprocess_image(image_path)
        feature = model.predict(image)
        features[image_path] = feature.flatten()  # Flatten the feature vector
    return features

# Find Similar Images
def find_similar_images(query_image_path, features, model, top_n=5):
    query_image = preprocess_image(query_image_path)
    query_feature = model.predict(query_image).flatten()
    
    similarities = {}
    for image_path, feature in features.items():
        similarity = cosine_similarity([query_feature], [feature])[0][0]
        similarities[image_path] = similarity

    # Sort by similarity
    sorted_similarities = sorted(similarities.items(), key=lambda x: x[1], reverse=True)
    return sorted_similarities[:top_n]


dataset_dir = "data/Apparel/Boys/"
image_paths = [os.path.join(dataset_dir, fname) for fname in os.listdir(dataset_dir) if fname.endswith(('.png', '.jpg', '.jpeg'))]

# Load the model
cnn_model = load_model()

# Extract features for the dataset and store them in a dictionary for future search
print("Extracting features from dataset...")
dataset_features = extract_features(image_paths, cnn_model)
print("Feature extraction completed!")
```

```python
# Query for similar images
query_image = "data/Apparel/Boys/34024.jpg" # Example Image
print(f"Finding images similar to {query_image}...")
similar_images = find_similar_images(query_image, dataset_features, cnn_model)

# Display Results
print("Top Similar Images:")
for image_path, similarity in similar_images:
    print(f"{image_path} - Similarity: {similarity:.4f}")
```

## 3. Siamese Network

```python
# Define Hyperparameters
IMG_SIZE = (100, 100, 3)
BATCH_SIZE = 32
EPOCHS = 20
MARGIN = 1.0  # Margin for contrastive loss

# Helper: Preprocess image
def preprocess_image(image_path):
    image = load_img(image_path, target_size=IMG_SIZE[:2])
    image_array = img_to_array(image)
    image_array = image_array / 255.0  # Normalize to [0, 1]
    return image_array

# Helper: Generate image pairs (randomly for example)
def generate_pairs(image_paths, labels, num_pairs=1000):
    pairs = []
    targets = []
    label_to_images = {}
    for path, label in zip(image_paths, labels):
        label_to_images.setdefault(label, []).append(path)
    
    for _ in range(num_pairs):
        label = random.choice(list(label_to_images.keys()))
        positive_pair = random.sample(label_to_images[label], 2)
        pairs.append((positive_pair[0], positive_pair[1]))
        targets.append(1)  # Positive pair
        
        # Negative pair
        neg_label = random.choice([l for l in label_to_images.keys() if l != label])
        neg_pair = (random.choice(label_to_images[label]), random.choice(label_to_images[neg_label]))
        pairs.append(neg_pair)
        targets.append(0)  # Negative pair

    return pairs, targets

# Define Siamese Network Base Model
def build_base_model():
    inputs = Input(shape=IMG_SIZE)
    x = Conv2D(64, (7, 7), activation='relu', padding='same')(inputs)
    x = MaxPooling2D()(x)
    x = Conv2D(128, (5, 5), activation='relu', padding='same')(x)
    x = MaxPooling2D()(x)
    x = Conv2D(256, (3, 3), activation='relu', padding='same')(x)
    x = Flatten()(x)
    x = Dense(512, activation='relu')(x)
    return Model(inputs, x, name="Siamese_Base")

# Define Siamese Network
def build_siamese_network(base_model):
    input_a = Input(shape=IMG_SIZE)
    input_b = Input(shape=IMG_SIZE)

    # Get embeddings
    embedding_a = base_model(input_a)
    embedding_b = base_model(input_b)

    # Compute L2 distance
    l2_distance = Lambda(lambda tensors: K.abs(tensors[0] - tensors[1]))([embedding_a, embedding_b])
    output = Dense(1, activation='sigmoid')(l2_distance)

    return Model([input_a, input_b], output)

# Contrastive Loss
def contrastive_loss(y_true, y_pred):
    y_true = tf.cast(y_true, dtype=tf.float32)  # Ensure y_true is float32
    squared_pred = tf.square(y_pred)
    squared_margin = tf.square(tf.maximum(MARGIN - y_pred, 0))
    return tf.reduce_mean(y_true * squared_pred + (1 - y_true) * squared_margin)
    
# Prepare Dataset
def prepare_data(image_dir):
    image_paths = []
    labels = []
    for label, folder in enumerate(os.listdir(image_dir)):
        folder_path = os.path.join(image_dir, folder)
        for image_name in os.listdir(folder_path):
            image_paths.append(os.path.join(folder_path, image_name))
            labels.append(label)
    return image_paths, labels


# Dataset path
dataset_dir = "data/Apparel/"
image_paths, labels = prepare_data(dataset_dir)

# # Generate pairs
print("Generating image pairs...")
pairs, targets = generate_pairs(image_paths, labels)
pairs = [(preprocess_image(p[0]), preprocess_image(p[1])) for p in tqdm(pairs)]
targets = np.array(targets)

# Split data into train and test
split_idx = int(0.8 * len(pairs))
train_pairs, test_pairs = pairs[:split_idx], pairs[split_idx:]
train_targets, test_targets = targets[:split_idx], targets[split_idx:]

# Build Siamese Network
print("Building Siamese Network...")
base_model = build_base_model()
siamese_network = build_siamese_network(base_model)
siamese_network.compile(optimizer='adam', loss=contrastive_loss, metrics=['accuracy'])

# Prepare train and test data
train_a = np.array([p[0] for p in train_pairs])
train_b = np.array([p[1] for p in train_pairs])
test_a = np.array([p[0] for p in test_pairs])
test_b = np.array([p[1] for p in test_pairs])

# Train the model
print("Training Siamese Network...")
siamese_network.fit(
    [train_a, train_b], train_targets,
    validation_data=([test_a, test_b], test_targets),
    batch_size=BATCH_SIZE,
    epochs=EPOCHS
)
```

## 4. Vision Tranformer (ViTs)

**Vision transformer is samilar to CNN, the key concept is ViTs break down an image into patches, serializes each patch into a vector, and then processes the vectors with a transformer encoder.**


```python


# Parameters
image_size = 128  # Image dimensions (128x128)
patch_size = 16   # Patch size (16x16)
num_patches = (image_size // patch_size) ** 2
embed_dim = 64    # Embedding size
num_heads = 4     # Number of attention heads
transformer_layers = 4
mlp_dim = 128     # Hidden size of the MLP

# Prepare Image Patches
def create_patches(image, patch_size):
    patches = tf.image.extract_patches(
        images=tf.expand_dims(image, axis=0),
        sizes=[1, patch_size, patch_size, 1],
        strides=[1, patch_size, patch_size, 1],
        rates=[1, 1, 1, 1],
        padding="VALID",
    )
    patches = tf.reshape(patches, [-1, patch_size * patch_size * 3])
    return patches

# Positional Embeddings
class PositionalEmbedding(tf.keras.layers.Layer):
    def __init__(self, num_patches, embed_dim):
        super(PositionalEmbedding, self).__init__()
        # Positional embeddings for each patch
        self.pos_embedding = tf.Variable(
            tf.random.normal([1, num_patches, embed_dim]), trainable=True
        )

    def call(self, x):
        return x + self.pos_embedding

# MLP Layer
def mlp(x, hidden_units, dropout_rate):
    for units in hidden_units:
        x = Dense(units, activation=tf.nn.gelu)(x)
        x = Dropout(dropout_rate)(x)
    return x

# Vision Transformer Encoder
def vision_transformer(num_patches, embed_dim, num_heads, transformer_layers, mlp_dim):
    inputs = Input(shape=(num_patches, embed_dim))
    x = inputs

    for _ in range(transformer_layers):
        # Multi-Head Attention
        x1 = LayerNormalization(epsilon=1e-6)(x)
        attention_output = MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)(x1, x1)
        x2 = Add()([x, attention_output])

        # MLP
        x3 = LayerNormalization(epsilon=1e-6)(x2)
        x3 = mlp(x3, [mlp_dim, embed_dim], 0.1)
        x = Add()([x2, x3])

    outputs = LayerNormalization(epsilon=1e-6)(x)
    model = Model(inputs, outputs)
    return model

# Image Feature Extractor
def build_vit_model(image_size, patch_size, embed_dim, num_heads, transformer_layers, mlp_dim):
    num_patches = (image_size // patch_size) ** 2
    patch_dim = patch_size * patch_size * 3

    # Input Layer
    inputs = Input(shape=(image_size, image_size, 3))

    # Create Patches
    patches = tf.image.extract_patches(
        images=inputs,
        sizes=[1, patch_size, patch_size, 1],
        strides=[1, patch_size, patch_size, 1],
        rates=[1, 1, 1, 1],
        padding="VALID"
    )
    patch_embeddings = Dense(embed_dim)(patches)

    # Flatten Patch Grid for Positional Embeddings
    patch_embeddings = tf.reshape(patch_embeddings, (-1, num_patches, embed_dim))

    # Add Positional Embeddings
    pos_embeddings = PositionalEmbedding(num_patches, embed_dim)(patch_embeddings)

    # Transformer Encoder
    transformer = vision_transformer(num_patches, embed_dim, num_heads, transformer_layers, mlp_dim)
    transformer_output = transformer(pos_embeddings)

    # Feature Aggregation
    features = Flatten()(transformer_output)
    model = Model(inputs, features)
    return model

# Build the ViT model
vit_model = build_vit_model(image_size, patch_size, embed_dim, num_heads, transformer_layers, mlp_dim)
vit_model.summary()

# Load Sample Images
def load_image(path, size):
    img = tf.keras.preprocessing.image.load_img(path, target_size=(size, size))
    img = tf.keras.preprocessing.image.img_to_array(img) / 255.0
    return img

image1 = load_image("data/Apparel/Boys/15574.jpg", image_size)
image2 = load_image("data/Footwear/Men/10266.jpg", image_size)

# Extract Features
features1 = vit_model.predict(np.expand_dims(image1, axis=0))
features2 = vit_model.predict(np.expand_dims(image2, axis=0))

# Compute Similarity
similarity = cosine_similarity(features1, features2)
print(f"Image Similarity: {similarity[0][0]:.4f}")
```

## 5. Autoencoder Approach

```python


# Parameters
image_size = 64
latent_dim = 64

# Autoencoder Model
def build_autoencoder(image_size, latent_dim):
    # Encoder
    input_img = Input(shape=(image_size, image_size, 3))
    x = Flatten()(input_img)
    encoded = Dense(latent_dim, activation='relu')(x)

    # Decoder
    x = Dense(image_size * image_size * 3, activation='sigmoid')(encoded)
    decoded = Reshape((image_size, image_size, 3))(x)

    # Models
    autoencoder = Model(input_img, decoded, name="Autoencoder")
    encoder = Model(input_img, encoded, name="Encoder")
    return autoencoder, encoder

# Build Autoencoder
autoencoder, encoder = build_autoencoder(image_size, latent_dim)
autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

# Example Dataset (Using CIFAR-10 for demonstration)
(x_train, _), (x_test, _) = tf.keras.datasets.cifar10.load_data()
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# Resize to target image size
x_train = tf.image.resize(x_train, (image_size, image_size))
x_test = tf.image.resize(x_test, (image_size, image_size))

# Train Autoencoder
autoencoder.fit(x_train, x_train, epochs=10, batch_size=64, validation_data=(x_test, x_test))

# Extract Features
def extract_features(encoder, images):
    return encoder.predict(images)

# Compute Similarity
def compute_similarity(features1, features2):
    return cosine_similarity(features1, features2)

# Test Similarity
image1 = x_test[0]
image2 = x_test[1]
image3 = x_test[10]

# Extract Features
features1 = extract_features(encoder, np.expand_dims(image1, axis=0))
features2 = extract_features(encoder, np.expand_dims(image2, axis=0))
features3 = extract_features(encoder, np.expand_dims(image3, axis=0))

# Compute Similarity
similarity_12 = compute_similarity(features1, features2)
similarity_13 = compute_similarity(features1, features3)

print(f"Similarity between Image 1 and Image 2: {similarity_12[0][0]:.4f}")
print(f"Similarity between Image 1 and Image 3: {similarity_13[0][0]:.4f}")

# Visualize Results
plt.figure(figsize=(6, 2))
plt.subplot(1, 3, 1)
plt.title("Image 1")
plt.imshow(image1)
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Image 2")
plt.imshow(image2)
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Image 3")
plt.imshow(image3)
plt.axis("off")
plt.show()
```

1. Hashing Method:

		Advantages: Extremely fast since it relies on linear search or dictionary lookups.
		Drawbacks: Very low precision and recall because hashing algorithms are sensitive to minor changes in the image (e.g., rotation or scaling).
		Best Use Case: Simple or static datasets.

2. CNN-Based Feature Extraction:

		Advantages: Extracts robust features from images, resulting in good similarity performance.
		Drawbacks: May miss some finer-grained similarities unless trained on a domain-specific dataset.
		Best Use Case: General-purpose image similarity tasks.

3. Siamese Network:

		Advantages: Specifically designed for similarity learning, with very high precision and recall.
		Drawbacks: Requires labeled pairs for training, which can be resource-intensive.
		Best Use Case: Scenarios needing pairwise similarity comparison (e.g., face verification).

4. Vision Transformer (ViT):

		Advantages: State-of-the-art performance with high precision, recall, and retrieval accuracy, especially on diverse datasets.
		Drawbacks: Computationally intensive, requiring powerful hardware for training and inference.
		Best Use Case: Large and diverse datasets requiring fine-grained similarity detection.

5. Autoencoder-Based Approach:

		Advantages: Works well for unsupervised feature extraction and can handle reconstruction tasks effectively.
		Drawbacks: Moderate precision and recall, as its not explicitly designed for similarity tasks.
		Best Use Case: Use cases where reconstruction is a primary concern alongside similarity.

| **Approach**            | **Precision** | **Recall** | **Retrieval Accuracy** | **Efficiency**       | **Scalability**    |
|--------------------------|---------------|------------|-------------------------|----------------------|--------------------|
| **Hashing Method**       | Low           | Low        | Low                    | Very High (Fast)     | Low                |
| **CNN-Based Feature Extraction** | High  | Medium     | High                   | Moderate             | Moderate           |
| **Siamese Network**      | Very High     | High       | Very High              | Moderate to High     | Moderate           |
| **Vision Transformer (ViT)** | Very High | Very High  | Very High              | High (on GPU)        | High               |
| **Autoencoder-Based**    | Medium        | Medium     | Medium                 | Moderate             | High               |

## Fine tuning Dataset

Fine-tuning the dataset involves preprocessing or curating the dataset to better align with the requirements of the task. For an image similarity search, fine-tuning the dataset means:

-	Balancing Classes: Ensuring that the dataset has a balanced number of samples per class (if applicable).
-	Data Augmentation: Generating variations of the images to increase the dataset size and diversity.
-	Normalization: Standardizing pixel values to a specific range (e.g., 0-1 or -1 to 1).
-	Removing Outliers: Eliminating mislabeled or noisy data.
-	Consistent Resolution: Resizing all images to the same dimensions.
-	Splitting the Dataset: Dividing the dataset into training, validation, and testing subsets.
